```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
data = sns.load_dataset("iris")
```

```python
data.head(5)
```

```python
data['species'].nunique()
```

```python
data['species'].unique()
```

```python
sns.kdeplot(data=data, x="sepal_length", fill=True)
```

```python
sns.kdeplot(data=data, x="sepal_length", fill=True)
plt.title("Univariate Distribution")
plt.show()
```

```python
sns.kdeplot(data=data, x="sepal_length", fill=False)
```

```python
sns.kdeplot(
    data=data,
    x="sepal_length",
    y="sepal_width",
    fill=True
)
plt.title("Bivariate Distibution")
plt.show()
```

```python
sns.jointplot(
    data=data,
    x="sepal_length",
    y="sepal_width",
    kind="scatter"
)
```

```python
sns.pairplot(data)
plt.show()
```
