```python
# 8] Multilevel Inheritance
```

```python
class GrandParent:
    def gp(self):
        print("GrandParent")

class Parent(GrandParent):
    def p(self):
        print("Parent")

class Child(Parent):
    def c(self):
        print("Child")

obj = Child()

obj.gp()
obj.p()
obj.c()
```

```python
# 9] Multiple Inheritance
```

```python
# 12] Demonstrate Encapsulation using private variables, getter methods, and setter methods.
```

```python
class Student:
    def __init__(self):
        self.__marks = 0      # Private Variable

    # Setter Method
    def set_marks(self, marks):
        self.__marks = marks

    # Getter Method
    def get_marks(self):
        return self.__marks

s = Student()

s.set_marks(95)

print("Student Marks:", s.get_marks())
```

```python
# 13] Implement Polymorphism using Method Overriding.
```

```python
class Bird:
    def fly(self):
        print("Bird can fly")

class Sparrow(Bird):
    def fly(self):
        print("Sparrow flies high")

class Penguin(Bird):
    def fly(self):
        print("Penguin cannot fly")

b = Bird()
s = Sparrow()
p = Penguin()

b.fly()
s.fly()
p.fly()
```

```python
# 14] Create an Abstract Class, implement abstract methods, and instantiate child classes.
```

```python
from abc import ABC, abstractmethod

class Shape(ABC):

    @abstractmethod
    def area(self):
        pass

class Square(Shape):
    def __init__(self, side):
        self.side = side

    def area(self):
        print("Area of Square =", self.side * self.side)

s = Square(4)
s.area()
```

```python
# 15] Demonstrate Abstraction using the abc module with multiple child classes.
```

```python
from abc import ABC, abstractmethod

class Vehicle(ABC):

    @abstractmethod
    def start(self):
        pass

class Car(Vehicle):
    def start(self):
        print("Car starts with a key.")

class Bike(Vehicle):
    def start(self):
        print("Bike starts with a self-start button.")

c = Car()
b = Bike()

c.start()
b.start()
```

```python
# 16] Write a Python program to overload arithmetic operators, compare objects, and display the results.
```

```python
class Number:
    def __init__(self, value):
        self.value = value

    # Overloading + operator
    def __add__(self, other):
        return self.value + other.value

    # Overloading > operator
    def __gt__(self, other):
        return self.value > other.value

n1 = Number(20)
n2 = Number(10)

print("Addition =", n1 + n2)
print("n1 > n2 =", n1 > n2)
```

```python
# 18] Write a Python program to demonstrate object composition, access composed objects, and display their details.
```

```python
class Engine:
    def __init__(self, engine_type):
        self.engine_type = engine_type

    def display(self):
        print("Engine Type:", self.engine_type)

class Car:
    def __init__(self, brand, engine):
        self.brand = brand
        self.engine = engine

    def display(self):
        print("Car Brand:", self.brand)
        self.engine.display()

e = Engine("Petrol")
c = Car("Toyota", e)

c.display()
```

```python
# 24] Write a Python program to handle exceptions using try, except, else, and finally.
```

```python
try:
    a = int(input("Enter first number: "))
    b = int(input("Enter second number: "))

    result = a / b

except ZeroDivisionError:
    print("Division by zero is not allowed.")

except ValueError:
    print("Please enter valid numbers.")

else:
    print("Result =", result)

finally:
    print("Program Finished.")
```

```python
# 25] Write a Python program to create a simple Student Management System using OOP.
```

```python
class Student:
    def __init__(self, roll, name, marks):
        self.roll = roll
        self.name = name
        self.marks = marks

    def display(self):
        print("Roll No :", self.roll)
        print("Name    :", self.name)
        print("Marks   :", self.marks)


students = []

while True:
    print("\n1. Add Student")
    print("2. Display Students")
    print("3. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        roll = input("Enter Roll No: ")
        name = input("Enter Name: ")
        marks = input("Enter Marks: ")

        s = Student(roll, name, marks)
        students.append(s)

        print("Student Added Successfully!")

    elif choice == "2":
        if len(students) == 0:
            print("No Student Records Found.")
        else:
            print("\nStudent Records")
            for s in students:
                s.display()
                print("--------------------")

    elif choice == "3":
        print("Thank You!")
        break

    else:
        print("Invalid Choice")
```
