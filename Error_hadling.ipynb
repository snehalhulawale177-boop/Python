```python
# Try-except
```

```python
# Try contains code that might cause an error if an error happens, except handless it.
```

```python
try:
    x=10 / 0
except ZeroDivisionError:
    print("cannot divide by zero")
```

```python
# Multiple except blocks
# Different errors can be handled in different ways using multiple except blocks.
```

```python
try:
    num = int("abc")
    result = 10 / num
except ValueError:
    print("conversion failed")
except ZeroDivisionError:
    print("Diision by zero error")
```

```python
try:
    x=10 / 0
    result = 10 / num
except ValueError:
    print("conversion failed")
except ZeroDivisionError:
    print("Division by zero error")
```

```python
# 3.exception in one except
# you can handled more than one error type in a single block
```

```python
try:
    x=int("abc")
except(ValueError,TypeError):
    print("Value or Type error occurred")
```

```python
# 4.Generic Exception
# Catches any kind of error when you don't specify the exact type.
```

```python
try:
    x=int("abc")
except Exception as e:
    print("Error occurred:",e)
```

```python
# 5. Else block
# Runs only when no error occurs in the try block.
```

```python
try:
    x = 10 / 2
except ZeroDivivisionError:
    print("Error")
else:
    print("Success",x)
```

```python
# finally block
# Runs every time, Whether an error happens or not.often used for cleanup.
```

```python
try:
    x=10 / 0
except ZeroDivisionError:
    print("Error occurred")
finally:
    print("This always runs")
```
