```python
# 2D data structures in Pandas
```

```python
data = {
    "Customer_ID": [101, 102, 103, 104, 105],
    "Name": ["Snehal", "Tanishka", "Savi", "Sanika", "Rani"],
    "Age": [18, 25, None, 20, 30],
    "City": ["Pune", "Mumbai", "Pune", "Delhi", "Mumbai"],
    "Salary": [50000, 60000, 55000, None, 65000]
}
```

```python
df = pd.DataFrame(data)
```

```python
print(df)
```

```python
# Convert DataFrame to CSV
```

```python
df.to_csv("customer_data.csv", index=False)
```

```python
# Load Dataset from CSV
```

```python
df = pd.read_csv("customer_data.csv")
```

```python
df
```

```python
# Handling Missing Values
```

```python
df.isnull().sum()
```

```python
# Fill missing values
```

```python
# Fill Age with mean
```

```python
df["Age"] = df["Age"].fillna(df["Age"].mean())
```

```python
df
```

```python
# Fill Salary with median
```

```python
df["Salary"] = df["Salary"].fillna(df["Salary"].median())
```

```python
# Data Cleaning
```

```python
df = df.drop_duplicates()
```

```python
df.rename(columns={"Name": "Customer_Name"}, inplace=True)
```

```python
df
```

```python
# Change datatype
```

```python
df["Age"] = df["Age"].astype(int)
```

```python
# Filtering Data
```

```python
# Customers with Salary > 55000
```

```python
high_salary = df[df["Salary"] > 55000]
print(high_salary)
```

```python
# Selecting Data
```

```python
# Single column
```

```python
print(df["Customer_Name"])
```

```python
# Multiple columns
```

```python
print(df[["Customer_Name", "Salary"]])
```

```python
# Row selection
```

```python
print(df.iloc[0])  # first row
```

```python
# Sorting Data
```

```python
df = df.sort_values(by="Salary", ascending=False)
```

```python
df
```

```python
# Merge Data
```

```python
df1 = pd.DataFrame({
    "Customer_ID": [101, 102],
    "Product": ["Car", "Bike"]
})
```

```python
merged = pd.merge(df, df1, on="Customer_ID", how="left")
```
