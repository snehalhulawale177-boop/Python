```python
# 4] Numbers 1 to 50 and sum
```

```python
sum_even = 0

print("Even numbers from 1 to 50:")
for i in range (1,51):
    if i % 2 == 0:
        print(i,end=" ")
        sum_even +=i

print("\n\nsum of all even numbers =", sum_even)
```

```python
# 5] Multiplication Table
```

```python
num = int(input("Enter a number: "))
limit = int(input("Enter the limit:"))

count = 0

print("\nNumber\tsquare")

for i in range(1,limit+1):
    multiple = num * i
    square = multiple ** 2

    print(multiple, "\t", square)
    count +=1

print("\nTotal multiples generated = ",count)
    
```

```python
# 6] Factorial, Digits, Reverse 
```

```python
# 10] Count And Replace
```

```python
lst = [10,20,30,30,40,70]

x = int(input("Enter element to find: "))
y = int(input("Enter new value: "))

print("occurrence=",lst.count(x))

if x in lst:
    pos = lst.index(x)
    print("first positon =",pos)
    lst[pos] = y

print("update List = ",lst)
```
